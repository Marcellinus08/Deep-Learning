LOAD DATA & PEMBERSIHAN DATA

```python
df = pd.read_csv('RegresiUTSTelkom.csv')

df = df.dropna()
df = df.apply(pd.to_numeric, errors='coerce')
df = df.dropna()
df.reset_index(drop=True, inplace=True)
```

FITUR & TARGET

```python
# Regresi target: kolom terakhir
target_reg = df.columns[-1]
X_reg = df.drop(columns=[target_reg])
y_reg = df[target_reg]
```

```python
# Klasifikasi target: kolom pertama (2001)
target_clf = df.columns[0]
X_clf = df.drop(columns=[target_clf])
y_clf = df[target_clf]
```

CEK TIPE FITUR

```python
numerical_features = X_reg.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_reg.select_dtypes(include=['object']).columns.tolist()
```

PREPROCESSING

```python
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])
```

REGRESI

```python
# Preprocessing untuk regresi
Xr = preprocessor.fit_transform(X_reg)
Xr_train, Xr_test, yr_train, yr_test = train_test_split(Xr, y_reg, test_size=0.2, random_state=42)

# Membangun model regresi
model_reg = Sequential([
    Dense(256, activation='relu', input_shape=(Xr.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1)  # Output untuk regresi
])

# Kompilasi model dengan loss MSE
model_reg.compile(optimizer='adam', loss='mse')

# Training model dengan verbose=1 untuk melihat progres setiap epoch
model_reg.fit(Xr_train, yr_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)
```

EVALUASI REGRESI

```python
yr_pred = model_reg.predict(Xr_test).flatten()
mse = mean_squared_error(yr_test, yr_pred)
rmse = np.sqrt(mse)
r2 = r2_score(yr_test, yr_pred)

print("\n--- Evaluasi Regresi ---")
print(f"RMSE: {rmse:.4f}")
print(f"MSE: {mse:.4f}")
print(f"R²: {r2:.4f}")
```

VISUALISASI REGRESI

```python
plt.figure(figsize=(6,6))
plt.scatter(yr_test, yr_pred, alpha=0.5)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Predicted vs Actual (Regresi)')
plt.grid(True)
plt.show()
```

```python
# Memilih kolom target dan fitur (target: kolom pertama yaitu '2001')
target_column = '2001'  # Gantilah dengan kolom target yang sesuai
X_clf = df.drop(target_column, axis=1)  # Fitur
y_clf = df[target_column]  # Target

# Preprocessing fitur
numerical_features = X_clf.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X_clf.select_dtypes(include=['object']).columns.tolist()

# Preprocessing dengan scaling dan encoding
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
])

Xc = preprocessor.fit_transform(X_clf)

# Label Encoding target
le = LabelEncoder()
y_clf_encoded = le.fit_transform(y_clf)

# Split data menjadi train dan test
Xc_train, Xc_test, yc_train, yc_test = train_test_split(Xc, y_clf_encoded, test_size=0.2, random_state=42)
```

```python
# Model Klasifikasi
model_clf = Sequential([
    Dense(256, activation='relu', input_shape=(Xc.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(len(np.unique(yc_train)), activation='softmax')  # n_classes adalah jumlah kelas
])

# Kompilasi model
model_clf.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Latih model
model_clf.fit(Xc_train, yc_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.preprocessing import LabelBinarizer
import seaborn as sns
from sklearn.metrics import ConfusionMatrixDisplay

# Prediksi hasil pada data test
y_pred = model_clf.predict(Xc_test)

# Prediksi kelas (mengambil kelas dengan probabilitas tertinggi)
y_pred_classes = np.argmax(y_pred, axis=1)

# Hitung metrik evaluasi
accuracy = accuracy_score(yc_test, y_pred_classes)
precision = precision_score(yc_test, y_pred_classes, average='macro', zero_division=1)
recall = recall_score(yc_test, y_pred_classes, average='macro', zero_division=1)
f1 = f1_score(yc_test, y_pred_classes, average='macro')

# Ensure y_pred and yc_test_bin have the same number of classes for ROC AUC calculation
num_classes = len(np.unique(yc_test))  # Get the actual number of classes

# Adjust y_pred if necessary
if y_pred.shape[1] != num_classes:
    # Create a new array with the correct shape, filled with zeros
    y_pred_adjusted = np.zeros((y_pred.shape[0], num_classes))

    # Copy the probabilities from y_pred to the adjusted array
    for i in range(num_classes):
        if i < y_pred.shape[1]:  # Avoid index error if y_pred has fewer columns
            y_pred_adjusted[:, i] = y_pred[:, i]

    y_pred = y_pred_adjusted  # Use the adjusted array for ROC AUC

# Gunakan probabilitas untuk AUC-ROC
lb = LabelBinarizer()
yc_test_bin = lb.fit_transform(yc_test)

# Calculate ROC AUC score
roc_auc = roc_auc_score(yc_test_bin, y_pred, multi_class='ovr', average='macro')

# Tampilkan hasil evaluasi
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"AUC-ROC: {roc_auc:.4f}")

# Visualisasi Confusion Matrix
cm = confusion_matrix(yc_test, y_pred_classes)
cm_display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)

# Plot Confusion Matrix
cm_display.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```
